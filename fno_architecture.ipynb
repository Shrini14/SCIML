```python
import torch
import numpy as np
```

```python
a = torch.tensor([[5.0, 3.0]])
```

```python


P = torch.tensor([[1.0, 0.5],
                  [0.2, 1.0]])

v = torch.matmul(a, P)
```

```python
fourier = torch.fft.rfft(v)
```

```python
R = torch.tensor([[1.0, 0.0],
                  [0.0, 1.0]], dtype=torch.complex64)

fourier_out = torch.matmul(fourier, R)
```

```python
inv_fourier = torch.fft.irfft(fourier_out, n=v.shape[-1])
```

```python
W = torch.tensor([[0.5, 0.0],
                  [0.0, 0.5]])

local = torch.matmul(v, W)
```

```python
v_next = torch.relu(inv_fourier + local)
```

```python
# next layer
v = v_next
```

```python
Q = torch.tensor([[1.0],
                  [1.0]])

u = torch.matmul(v, Q)
```
